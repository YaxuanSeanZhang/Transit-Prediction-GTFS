```python
import numpy as np
import pandas as pd
```

```python
# Read the csv file into the pandas dataframe
df_route = pd.read_csv("C:/Users/yaxua/Desktop/GEOG5541/group/rawdata/12-3/routes.txt")
df_trip = pd.read_csv("C:/Users/yaxua/Desktop/GEOG5541/group/rawdata/12-3/trips.txt")
df_calendar = pd.read_csv("C:/Users/yaxua/Desktop/GEOG5541/group/rawdata/12-3/calendar.txt")
```

```python
# Merge two dataframes based on the key field, and follow the order of the lef dataframe
df_merge_route_trip = pd.merge (df_route, df_trip, on = "route_id", how = "left")
df_merge_route_trip_calendar = pd.merge (df_merge_route_trip, df_calendar, on = "service_id", how = "left")
df_merge_extract = df_merge_route_trip_calendar.loc[:,["route_id" , "route_short_name" , "route_long_name" , "trip_id" ,"service_id", "trip_headsign", "direction_id","start_date","end_date"]]
df_merge_extract
```

```python
# Save to csv file and turn off the index
df_merge_extract.to_csv("C:/Users/yaxua/Desktop/GEOG5541/group/data/merge_extract_12-3.csv", index = False)
```

```python
# Read the csv file into the pandas dataframe
df_stop_times = pd.read_csv("stop_times.txt")
df_stop_times
```

```python
# Merge two dataframes based on the key field, and follow the order of the lef dataframe
df_merge_route_trip_stop_times = pd.merge(df_merge_route_trip, df_stop_times, on ="trip_id", how = "left" )
df_merge_route_trip_stop_times
```

```python
# Read the csv file into the pandas dataframe
df_stop = pd.read_csv("stops.txt")
df_stop
```

```python
# Merge two dataframes based on the key field, and follow the order of the lef dataframe
df_merge_rout_trip_stop_times_stop = pd.merge(df_merge_route_trip_stop_times, df_stop, on = "stop_id", how = "left")
df_merge_rout_trip_stop_times_stop
```

```python
# Save as csv and turn off the index
df_merge_rout_trip_stop_times_stop.to_csv("Merge.csv", index = False)
```
